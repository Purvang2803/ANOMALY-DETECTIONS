### DBSCAN ANAMOLY DETECTION

```python
import pandas as pd
from sklearn.cluster import DBSCAN
%matplotlib inline  
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles 
```

```python
X,y = make_circles(n_samples=750,factor=0.3, noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbcan =DBSCAN(eps=0.10)
```

```python
dbcan.fit_predict(X)
```

```python
dbcan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbcan.labels_)
plt.show()
```

### ISOLATION FOREST ANOMALY DETECTION

```python
from sklearn.ensemble import IsolationForest
```

```python
X, y = make_circles(n_samples=750, factor=0.3, noise=0.1)
```

```python
iso_forest = IsolationForest(contamination=0.05, random_state=42)  # 5% anomalies
y_pred = iso_forest.fit_predict(X)
```

```python
y_pred
```

```python
anomalies = X[y_pred == -1]
normals = X[y_pred == 1]
```

```python
plt.scatter(normals[:, 0], normals[:, 1], c='blue', label='Normal')
plt.scatter(anomalies[:, 0], anomalies[:, 1], c='red', label='Anomaly', marker='x')
plt.title("Isolation Forest - Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
